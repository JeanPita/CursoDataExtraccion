##**Conexión a Api de Presupuesto Abierto:**

Utilidad: Obtener los datos del credito del Análisis Transversal con la posibilidad de filtrar una etiqueta en especifico



*   Atención a Personas con Discapacidad
*   Cambio Climático
*   Inversión en Políticas de Niñez y Adolescencia
*   Presupuesto Ambiental y Cambio Climático
*   Políticas de Género

Estas son las etiquetas utilizadas actualmente, puede que no haya informacion actualizada de 2024 para algunas etiquetas








```python
import requests
```

```python
from google.colab import userdata
user_key=userdata.get('PRESUPUESTO_KEY')
```

**Filtro dinamico en donde seleccionamos la etiqueta para la cual queremos extraer los datos**

```python
etiqueta_input = input("Ingrese el nombre de la etiqueta (Puede copiar y pegar de la lista mencionada arriba): ")
##   ETIQUETAS QUE SE PUEDEN UTILIZAR PARA COMPLETAR EL CAMPO
            #"Políticas de Género"
            #"Atención a Personas con Discapacidad"
            #"Presupuesto Ambiental y Cambio Climático"
            #"Cambio Climático"
            #"Inversión en Políticas de Niñez y Adolescencia"
```

**Codigo de extracción**

```python
url= 'https://www.presupuestoabierto.gob.ar/api/v1/transversal_financiero'
headers = {"Authorization": user_key, 'Content-Type': 'application/json'}

data = {
    "title": "Análisis transversal financiero",


    "columns": ["ejercicio_presupuestario",
    "etiqueta_id",
    "etiqueta_desc",
    "programa_id",
    "programa_desc",
    "finalidad_id",
    "finalidad_desc",
    "funcion_id",
    "funcion_desc",
    "fuente_financiamiento_id",
    "fuente_financiamiento_desc",
    "credito_inicial_ponderado",
		"credito_vigente_ponderado",
		"credito_ejecutado_ponderado"

    ],
    "ejercicios": [2020,2021,2022,2023],

    	"filters": [
		{
		"column": "etiqueta_desc",
		"operator": "equal",
		"value": etiqueta_input
		}
	]
}

response = requests.post(url, headers=headers, json=data)
```

```python
response.text
```

```python
import pandas as pd
from io import StringIO

df = pd.read_csv(StringIO(response.text))
```
